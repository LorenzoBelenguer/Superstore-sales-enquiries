```python
# Which products were sold the most in the different US regions?

count = df_copy.groupby(['Category', 'Sub-Category', 'Region']).size().reset_index(name='Count')

print(count)
```

```python
# plotting the outcome

# Create a bar chart
plt.bar(count['Category'] + count['Sub-Category'] +['Region'], count['Count'])
plt.xlabel('Category - Subcategory')
plt.ylabel('Count')
plt.title('Count of Category - Subcategory Combinations')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.show()



```

```python
# Which States sold the most?
# California = 1997

count2 = df_copy["State"].value_counts()

print(count2)
```

```python
# plotting the outcome

# Create a bar chart
count2.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Which state sold the most')

plt.show()
```

```python
# How many products were sold each year
# We need to create a year column first with the last 4 digits form the date column

# Extract the year from the "Ship Date" column
df_copy['year'] = df_copy['Ship Date'].str[-4:]

print(df_copy)

```

```python
# How many products were sold each year

count3 = df_copy.groupby(['Sub-Category', 'year']).size().reset_index(name='Count')

print(count3)
```

```python
# plotting the outcome

# Create a bar chart
count3.plot(kind='bar')
plt.xlabel('Sold product per year')
plt.ylabel('Count')
plt.title('How many products were sold each year')

plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability

plt.show()
```

```python
# Most sold products each year


# Group by Year and calculate the value counts for each Product
grouped = df_copy.groupby('year')['Sub-Category'].value_counts()

# Find the most sold product for each year
count4 = grouped.groupby(level=0).idxmax().reset_index()

print(count4)
```

```python
# Which customer ordered the most and what was the total value of their purchases?


# Group by Customer and calculate the total value of purchases
grouped = df_copy.groupby('Customer Name')['Sales'].agg(['sum', 'count'])

# Find the customer with the highest total purchase value
count5 = grouped[grouped['sum'] == grouped['sum'].max()]


print(count5)
```
